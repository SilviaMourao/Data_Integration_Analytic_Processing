```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('Localizacao.xlsx')
df2=pd.read_excel('Vizinhos.xlsx', sheet_name='LB_Eurovision')
dflocal=pd.DataFrame(columns=['IDLocalizacao','Pais'])
```

```python
df2.rename(columns = {'Name':'English Name'}, inplace = True)
df2.drop(columns=['Neighbour10', 'Neighbour9','Neighbour8', 'Neighbour7','Neighbour6', 'Neighbour5','Neighbour4', 'Neighbour3','Neighbour2','Neighbour1','Unnamed: 3', 'Unnamed: 14'])
```

```python
print(df2)
```

```python
result = pd.merge(df1, df2, on=["English Name"])
```

```python
dflocal['Pais']=result['English Name']
```

```python
dflocal['Regiao']=result['Region of the territory']
dflocal['Continente']=result['Continent of the territory']
```

```python
dflocal['IDLocalizacao']=dflocal.index+1
```

```python
dflocal['NumeroDeVizinhos']=result['NrNeighboursinEurope']
```

```python
print(dflocal)
```

```python
dflocal.to_csv("DimLocalizacao.csv", index=False)
```
